```python
import pandas as pd

def unique_words(column):
    unique_words = set()
    for entry in column:
        if pd.notna(entry):
            words = entry.split(';')
            unique_words.update(words)
    return unique_words
```

```python
unique_tags = unique_words(artists['tags_mb'])
```

```python
len(unique_tags)
```

```python
unique_tags_list = list(unique_tags)

tags = pd.DataFrame({'tagId': range(1, len(unique_tags_list) + 1),
                   'name': unique_tags_list})
```

```python
tags.head(15)
```

```python
tags.to_csv("tags.csv")
```

```python
tags_dict = dict(zip(tags['name'], tags['tagId']))
```

```python
df_expanded = artists['tags_mb'].str.split(';', expand=True).stack().reset_index(level=1, drop=True).rename('tag_names')

artists = artists.drop('tags_mb', axis=1).join(df_expanded)
```

```python
artists
```

```python
artists = artists.drop(['artist_lastfm','country_lastfm','tags_lastfm'], axis=1)
```

```python
artists[artists['tag_names'].isna()]
```

```python
artist_tags = artists[["mbid","tag_names"]]
```

```python
def get_value(key):
  if key in tags_dict:
    return int(tags_dict.get(key))
  else:
    return None
```

```python
artist_tags2 = artist_tags.copy(deep=True)
```

```python
artist_tags2['tagId'] = artist_tags2['tag_names'].apply(get_value)
```

```python
artist_tags2
```

```python
artist_tags2 = artist_tags2[artist_tags2['tagId'].notna()]
```

```python
artist_tags2['tagId'] = artist_tags2['tagId'].apply(lambda x: int(x) if pd.notna(x) else x)
```

```python
artist_tags2.head()
```

```python
artist_tags2 = artist_tags2.drop('tag_names', axis=1)
```

```python
artist_tags2.to_csv("artistTags.csv")
```
