```python
def add_pid_trackid(df):
    track_id = 0
    for index, row in df.iterrows():
        pid = row['pid']
        json_array = row['tracks']
        for json_dict in json_array:
            json_dict['pid'] = pid
            json_dict['trackId'] = track_id
            track_id += 1
```

```python
add_pid_trackid(data_track_playlist_sample)
```

```python
artists = pd.read_csv("artists.csv")
```

```python
artists.head()
```

```python
artist_dict = artists.set_index('artist_mb')['mbid'].to_dict()
```

```python
artist_dict['Ashanti']
```

```python
def add_artist_id(row):
    json_array = row['tracks']
    for json_dict in json_array:
        artist_name = json_dict['artist_name']
        artist_id = artist_dict[artist_name] if artist_name in artist_dict else None
        json_dict['artistId'] = artist_id
    return json_array
```

```python
data_track_playlist_sample['tracks'] = data_track_playlist_sample.apply(add_artist_id, axis=1)
```

```python
data_track_playlist_sample_exploded = data_track_playlist_sample.explode('tracks', ignore_index=True)[["tracks"]]
```

```python
data_track_playlist_sample_exploded
```

```python
normalized_df = pd.concat([json_normalize(x) for x in data_track_playlist_sample_exploded['tracks']], ignore_index=True)
```
